# Importando as bibliotecas

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
```

### Lendo o arquivo DataFrame e fazendo a copia.

```python
df = pd.read_csv("car_prices.csv")
data = df.copy()
```

```python
carros_2010 = data.loc[lambda dados: dados["year"] == 2010] 
```

```python
quantidade = carros_2010['make'].value_counts().head()
top_5_marcas = quantidade.reset_index()
top_5_marcas.columns = ['make', 'quantidade']
```

# Grafico

```python
f, ax = plt.subplots(figsize=(15, 5))

sns.barplot(data=top_5_marcas, x='make', y='quantidade', 
            color="#19758C", edgecolor='#1FB0DE', linewidth=2)
ax.bar_label(ax.containers[0], padding=0.3, fontsize=14, color="white")
ax.set_title("Top 5 Marcas Mais Vendidos 2010", fontsize=24)
ax.set_xlabel("Marcas", fontsize=18)
ax.set_ylabel("Quantidade", fontsize=18)
ax.set_ylim((0, top_5_marcas['quantidade'].max() + 1000))
ax.set_facecolor('#4A4A4A')

plt.show()
```

# Segunda atividade

```python
filtro = carros_2010.loc[carros_2010["make"] == "Ford"]

quantidade_bodys_ford = filtro['body'].value_counts().head(3).reset_index()
quantidade_bodys_ford.columns = ['body', 'quantidade']

f, ax = plt.subplots(figsize=(15, 5))
sns.barplot(data=quantidade_bodys_ford, x='body', y='quantidade', 
            color="#19758C", edgecolor='#1FB0DE', linewidth=2) 
ax.bar_label(ax.containers[0], padding=0.3, fontsize=14, color="white")
ax.set_title("Top 3 Bodys Mais Vendidos da Ford", fontsize=24)
ax.set_xlabel("Bodys", fontsize=18)
ax.set_ylabel("Quantidade", fontsize=18)
ax.set_ylim((0, quantidade_bodys_ford['quantidade'].max() + 1000))
ax.set_facecolor('#4A4A4A')

plt.show()
```
