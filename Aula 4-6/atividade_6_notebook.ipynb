# Exercício 1

## Crie um programa que:

### - Receba um número do usuário;
### - Apresente a tabuada desse número.

## O Output deve ser similar ao exemplo abaixo:



```python
# Campo para inserir o valor que o usuario deseja ver a tabuada.
numero = int(input('Digite o valor:\n\n '))

# Calculo e gerandor da tabela.
for i in range(1,11):
    print(f'| {numero} X {i} = {numero * i}')
```

# Exercício 2

## Crie um programa que seja um somador de despesas simples:

### - Receba a quantidade de despesas;
### - Faça a soma de cada despesa em ma variável;
### - Ao final, apresente o valor total da despesa.

## O output deve ser similar ao exemplo abaixo:



```python
# Define as variaveis necessarias e campo para digitação do usuario.
qtd_despesas = int(input('Digite a quantidade de dispesas: '))
total = 0

# Define um loop de acordo com a quantidade de dispesas que o usuario teve.
for i in range(qtd_despesas):
    valor = float(input('Digite o valor da compra: '))
    total += valor

# Imprime na tela o valor total de todos os valores que o usuario gastou.
print(f'Você gastou um total de R${total:.2f}')
```

##### Exercício 3

## Crie um sistema de banco simples, com as seguintes opções:

### - Menu com 1: Depósito, 2: Saque e s: Sair
### -  A aplicação deve parar quando o cliente escolher 's'
### - Ao final, imprima uma mensagem com o saldo final do cliente.

```python
# Cração das duas principais variaveis com que fara que o sistema funcione.
saldo = 0
resposta = 0

# Inicia a criação do loop para somente ser interrompido se o valor na resposta for 4.
while resposta != 4:

    # Imprime na tela a parte index d sistema.
    print(' \n-------- Banco do Brasil --------')
    print(' |    [1] Depósito     [2] Saque   |')
    print(' |    [3] Saldo        [4] Sair    |')
    print(' -----------------------------------')

    # Campo para a digitação do comando.
    resposta = int(input('Comando: \n'))

    # inicia as funcionalidades usando o if como controle de fluxo.
    if resposta == 1:     # Controle de fluxo para o comando 1 depostio.
        val_deposito = float(input('Insira o valor que deseja depositar: '))
        saldo += val_deposito
        print(f'Você depositou R${val_deposito:.2f}. Seu novo saldo é R${saldo:.2f}.')
    elif resposta == 2:     # Controle de fluxo para o comando 2 saque.
        val_saque = float(input('Insira o valor que deseja sacar: '))
        if val_saque <= saldo:    # Caso o usuario tente sacar um valor maior do que disponivel.
            saldo -= val_saque
            print(f'Você sacou R${val_saque:.2f}. Seu novo saldo é R${saldo:.2f}.')
        else:
            print('Saldo insuficiente para realizar o saque.')
    elif resposta == 3:     # Controle de fluxo para o comando 3 saldo.
        print(f'Seu saldo atualmente é de R${saldo:.2f}')
    elif resposta == 4:    # Controle de fluxo para o comando 4 sair.
        print('Obrigado! Por hoje é só.')
        break

    # Caso o usuario digitar um valor invalido.
    else:
        print('Escolha inválida!')
```
