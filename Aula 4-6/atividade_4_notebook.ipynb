# Exerc√≠cio 1

### Sabendo que:



### Crie um programa que receba o valor do raio de uma circunfer√™ncia e imprima:

### * Use pi = 3.14









```python
pi = 3.14 # Resultado do PI
```

#### - C√°lculo da √°rea: 

```python
raio = float(input("Digite o valor do raio da circunferencia:")) # Pega o raio da circunferencia
diametro = 2* raio # Fazendo o calculo de raio com base no diametro
```

#### - C√°lculo do per√≠metro:


```python
perimetro = 2* pi * raio # Calculando o perimetro.
```

#### - C√°lculo do di√¢metro:

```python
area = pi* (raio**2) # Calculando area.

# Imprimindo os valores na tela

print(f"diametro:{diametro}")
print(f"perimetro:{perimetro}")
print(f"area:{area}")
```

# Exerc√≠cio 2

### Crie um c√≥digo que receba a temperatura em ¬∞Celcius e:


#### - Calcule a temperatura correspondente em ¬∞Farenheit

# ¬∞F = (¬∞C x 1.8) + 32

```python
temperatura_c = float(input('Digite a temperatura.(em ¬∞Celcius)\n')) # Pega a temperatura em celcius.
conversao_f = temperatura_c * 1.8 + 32 # Calcula e usa a formula para converter para farenheit.
print(f'Convertido para ¬∞Farenheit:\n [ ¬∞{conversao_f} Farenheits ]') # Imprime na tela o resultado da convers√£o.
```

#### - Calcule a temperatura correspondente em Kelvin

# K = ¬∞C + 273,15

```python
temperatura_celcius = float(input('Digite a temperatura.(em ¬∞Celcius)\n')) # Pega a temperatura em celcius.
conversao_k = temperatura_celcius + 273.15 # Calcula e usa a formula para converter para kelvin.
print(f'Convertido para Kelvin:\n [ ¬∞{conversao_k} Kelvins ]') # Imprime na tela o resultado da convers√£o.
```

# Exerc√≠cio 3

### Crie um programa que receba: 

#### as notas dos 4 bimestres de um aluno:

```python
bimestre1 = 8 # Pega a nota do primeiro bimestre.
bimestre2 = 9 # Pega a nota do segundo bimestre.
bimestre3 = 7.5 # Pega a nota do terceiro bimestre.
bimestre4 = 10 # Pega a nota do quarto bimestre.
```

#### Calcule a m√©dia das notas desse aluno:

```python
media = bimestre1 + bimestre2 + bimestre3 + bimestre4 # Soma todos os valores.
media = media / 4 # Pega o valor somado e divide pela quantidade de valores.
```

#### Printe uma mensagem com a m√©dia final desse aluno:

```python
print(f'A media das suas notas √© [{media:.1f}]') # Imprime na tela o resultado da divis√£o.
```
